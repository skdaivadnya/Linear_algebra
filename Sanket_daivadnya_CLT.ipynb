```python
import numpy as np
import seaborn as sns
import statistics as stat

# to simulate a normally distributed population

x = np.random.normal(size=10000)
sns.displot(x)
```

```python
sns.displot(x)
```

```python
sns.displot(x, kde=False)
```

```python
x_sample = np.random.choice(x, size=20, replace=False)
x_sample
```

```python
stat.mean(x_sample)
```

```python
def sample_mean_calculation(population_array, sample_size, n_samples):
  sample_means = []
  for i in range(n_samples):
    sample = np.random.choice(population_array, size=sample_size, replace=False)
    sample_mean = stat.mean(sample)
    sample_means.append(sample_mean)
  return sample_means
```

```python
_=sns.displot(sample_mean_calculation(x,10,10))
```

```python
_ = sns.displot(sample_mean_calculation(x, 10, 1000))
```

```python
_ = sns.displot(sample_mean_calculation(x, 1000, 1000))
```

```python
#sampling from skewed population

from scipy.stats import skewnorm
s=skewnorm.rvs(12, size=10000)
sns.displot(s)
```

```python
_ = sns.displot(sample_mean_calculation(s, 10, 1000))
```

```python
_ = sns.displot(sample_mean_calculation(s, 1000, 1000))
```

Sampling from a multimodal distribution



```python
m = np.concatenate((np.random.normal(size=10000), np.random.normal(loc = 4.0, size=10000)))
_ = sns.displot(m)
```

```python
_ = sns.displot(sample_mean_calculation(m, 900, 1000))
```

```python
#Sampling means from any distribution produces a normal sampling distribution


u = np.random.uniform(size=10000)
_ = sns.displot(u, kde=False)
```

```python
_ = sns.displot(sample_mean_calculation(u, 2000, 2000))
```

Hence, as the amount of sampling size increases we can observe that the sampling distribution of means will get normally distributed. Which helps us to drive statistical tests which are designed for normal distributions.

T-test an example of statistical test, helps us to check whether two samples come from different populations.
By using this test we can conclude that for the unknown distribution of population, Central limit theorem can help us produce exceptional results.
